```python
def get_top_tags(data, tag_columns=['tag_1', 'tag_2', 'tag_3', 'tag_4', 'tag_5']):
    all_tags = []
    for col in tag_columns:
        all_tags.extend(data[col].dropna().values)
    return Counter(all_tags)

top_tags = get_top_tags(data)
top_15_tags = dict(sorted(top_tags.items(), key=lambda x: x[1], reverse=True)[:15])

plt.figure(figsize=(12, 6))
plt.bar(top_15_tags.keys(), top_15_tags.values())
plt.title('Top 15 Most Common Tags')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Data Preprocessing Steps

# 1. Handle missing values
def preprocess_data(df):
    # Create a copy to avoid modifying original data
    df = df.copy()
    
    # Fill missing values in text columns
    text_columns = ['subject', 'body', 'answer']
    for col in text_columns:
        df[col] = df[col].fillna('')
    
    # Fill missing values in tags with 'None'
    tag_columns = ['tag_1', 'tag_2', 'tag_3', 'tag_4', 'tag_5', 'tag_6', 'tag_7', 'tag_8']
    for col in tag_columns:
        df[col] = df[col].fillna('None')
    
    return df

# Apply preprocessing
data = preprocess_data(data_copy)

# 2. Text cleaning function
import re
def clean_text(text):
    # Convert to string in case of non-string input
    text = str(text)
    
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '<email>', text)
    
    # Remove phone numbers
    text = re.sub(r'[\+]?[(]?[0-9]{3}[)]?[-\s\.]?[0-9]{3}[-\s\.]?[0-9]{4,6}', '<phone>', text)
    
    # Remove account numbers (assuming they're in angle brackets)
    text = re.sub(r'<acc_num>', '<account>', text)
    text = re.sub(r'<tel_num>', '<phone>', text)
    text = re.sub(r'<ref_num>', '<reference>', text)
    
    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()

# 3. Apply text cleaning
data['clean_subject'] = data['subject'].apply(clean_text)
data['clean_body'] = data['body'].apply(clean_text)
data['clean_answer'] = data['answer'].apply(clean_text)

# 4. Create feature columns
data['text_length'] = data['clean_body'].str.len()
data['answer_length'] = data['clean_answer'].str.len()

# 5. Combine tags into a list
data['tags'] = data.apply(
    lambda x: [tag for tag in [x['tag_1'], x['tag_2'], x['tag_3'], x['tag_4'], x['tag_5']] 
              if tag != 'None'],
    axis=1
)

# 6. Create binary columns for high-priority and technical issues
data['is_high_priority'] = (data['priority'] == 'high').astype(int)
data['is_technical'] = data['tags'].apply(lambda x: 'Technical' in x).astype(int)

# 7. Print preprocessing summary
print("Preprocessing Summary:")
print("-" * 50)
print(f"Total samples: {len(data)}")
print(f"Missing values after preprocessing:\n{data.isnull().sum()}")
print("\nSample sizes by type:")
print(data['type'].value_counts())
print("\nSample sizes by language:")
print(data['language'].value_counts())
print("\nPriority distribution:")
print(data['priority'].value_counts())

# 8. Save preprocessed data
data.to_csv('preprocessed_tickets_dataset.csv', index=False)

# 9. Create feature matrices for modeling
from sklearn.preprocessing import LabelEncoder

# Encode categorical variables
le_type = LabelEncoder()
le_priority = LabelEncoder()
le_queue = LabelEncoder()
le_language = LabelEncoder()

data['type_encoded'] = le_type.fit_transform(data['type'])
data['priority_encoded'] = le_priority.fit_transform(data['priority'])
data['queue_encoded'] = le_queue.fit_transform(data['queue'])
data['language_encoded'] = le_language.fit_transform(data['language'])

# Display sample of preprocessed data
print("\nSample of preprocessed data:")
display(data[['clean_subject', 'clean_body', 'clean_answer', 'type', 'priority', 
             'language', 'tags', 'is_high_priority', 'is_technical']].head())

# 10. Basic statistics of preprocessed features
print("\nFeature Statistics:")
print("-" * 50)
print("\nText length statistics:")
print(data[['text_length', 'answer_length']].describe())

# 11. Correlation analysis
numeric_cols = ['text_length', 'answer_length', 'is_high_priority', 
                'is_technical', 'type_encoded', 'priority_encoded', 
                'queue_encoded', 'language_encoded']

plt.figure(figsize=(12, 8))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlations')
plt.tight_layout()
plt.show()
```
